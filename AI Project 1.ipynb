```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('data.csv')
```

```python
df
```

```python
df.info()
```

```python
df.drop(df.columns[[-1,0]],axis=1,inplace=True)
```

```python
df
```

```python
df['diagnosis'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
diag_map={'M':1,'B':0}
```

```python
df['diagnosis']=df['diagnosis'].map(diag_map)
```

```python
df
```

```python
x=df[['radius_mean','perimeter_mean','area_mean','concavity_mean','concave points_mean']]
```

```python
x
```

# Training with x

```python
y=df[['diagnosis']]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier()
```

```python
knn.fit(x_train,y_train)
```

```python
knn_y_pred=knn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

Accuracy Score with x dataset

```python
accuracy_score(knn_y_pred,y_test)
```

# Training with whole dataset df

```python
df.drop(df.columns[[0]],axis=1,inplace=True)
```

```python
x_train,x_test,y_train,y_test=train_test_split(df,y,test_size=0.2,random_state=42)
```

```python
knn.fit(x_train,y_train)
```

```python
knn_y_pred=knn.predict(x_test)
```

```python
accuracy_score(knn_y_pred,y_test)
```

We are getting better accuracy while selecting the whole dataset, so we will use whole df as the training set

# Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression(random_state=0)
```

```python
lr.fit(x_train,y_train)
```

```python
lr_y_pred=lr.predict(x_test)
```

```python
accuracy_score(lr_y_pred,y_test)
```

# Naive Bayes Model

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb=GaussianNB()
```

```python
gnb.fit(x_train,y_train)
```

```python
gnb_y_pred=gnb.predict(x_test)
```

```python
accuracy_score(gnb_y_pred,y_test)
```

# K-Fold Cross Validation

```python
from sklearn.model_selection import cross_val_score
```

```python
import warnings
warnings.filterwarnings("ignore")
```

Using KNN

```python
accuracy_all=[]
cvs_all=[]

scores=cross_val_score(knn,x,y,cv=10)

accuracy_all.append(accuracy_score(knn_y_pred,y_test))
cvs_all.append(np.mean(scores))

print("accuracy: {0:.2%}".format(accuracy_score(knn_y_pred,y_test)))
print("10-fold Cross Validation Score: {0:.2%} (+/-  {1:.2%})".format(np.mean(scores),np.std(scores)*2))
```

Using Logistic Regression

```python
accuracy_all=[]
cvs_all=[]

scores=cross_val_score(lr,x,y,cv=10)

accuracy_all.append(accuracy_score(lr_y_pred,y_test))
cvs_all.append(np.mean(scores))

print("accuracy: {0:.2%}".format(accuracy_score(lr_y_pred,y_test)))
print("10-fold Cross Validation Score: {0:.2%} (+/-  {1:.2%})".format(np.mean(scores),np.std(scores)*2))
```
